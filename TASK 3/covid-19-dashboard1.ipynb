```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt #Plotting
```

# Global Death Data

## Load Death Data

```python
df1 = pd.read_csv('/kaggle/input/covid19-data-from-john-hopkins-university/CONVENIENT_global_deaths.csv')
display('Deaths DF - Data SnapShot',df1.tail(3))
```

## PreProcessing

```python
# Sum total cases for each country and reset the index
df1a = df1.sum(axis=0).reset_index(name="Deaths")
df1a.columns = ["Country", "Deaths"]

# Remove the first row if necessary
df1a = df1a.iloc[1:, :]

# Convert 'Cases' column to numeric
df1a["Deaths"] = pd.to_numeric(df1a["Deaths"], errors="coerce")

# Filter rows where 'Cases' is an integer
df1a_final = df1a[df1a["Deaths"] % 1 == 0].copy()

# Sort the DataFrame by 'Cases' in descending order
df1a_final = df1a_final.sort_values("Deaths", ascending=False)

# Optionally, check the result
print(df1a_final.head())
```

```python
print(df1a_final.columns)
df1a_final.head()
```

## Top 10 Countries with highest death counts

```python
df1a_final.set_index('Country').head(10)
```

```python
print(df1a_final.columns)
```

## Death Plot

```python
# Ensure 'Country' is the index
df1a_final = df1a_final.set_index("Country")

# Select top 20 countries based on 'Deaths'
top_20 = df1a_final.nlargest(20, "Deaths")

# Plot
top_20.plot(
    kind="pie",
    y="Deaths",
    legend=False,
    ylabel="",
    title="Top 20 Countries by Deaths",
    fontsize=10,
    figsize=(10, 10),
    autopct='%1.1f%%'  # Show percentage on pie chart
)

plt.show()  # Display the plot
```

# Global Cases Data

```python
df2 = pd.read_csv('../input/covid19-data-from-john-hopkins-university/CONVENIENT_global_confirmed_cases.csv')
display('Cases DF - Data SnapShot',df1.tail(2))
```

```python
# Sum total cases for each country and reset the index
df2a = df2.sum(axis=0).reset_index(name="Cases")
df2a.columns = ["Country", "Cases"]

# Remove the first row if necessary
df2a = df2a.iloc[1:, :]

# Convert 'Cases' column to numeric
df2a["Cases"] = pd.to_numeric(df2a["Cases"], errors="coerce")

# Filter rows where 'Cases' is an integer
df2a_final = df2a[df2a["Cases"] % 1 == 0].copy()

# Sort the DataFrame by 'Cases' in descending order
df2a_final = df2a_final.sort_values("Cases", ascending=False)

# Optionally, check the result
print(df2a_final.head())
```

## Top 10 Countries with highest case count

```python
df2a_final.set_index('Country').head(10)
```

```python
print(df2a_final.columns)
```

```python
# Ensure 'Country' is the index
df2a_final = df2a_final.set_index("Country")

# Select top 20 countries based on 'Cases'
top_20_cases = df2a_final.nlargest(20, "Cases")

# Plot
top_20_cases.plot(
    kind="pie",
    y="Cases",
    legend=False,
    ylabel="",
    title="Top 20 Countries by Cases",
    fontsize=10,
    figsize=(10, 10),
    autopct='%1.1f%%'  # Show percentage on pie chart
)

plt.show()  # Display the plot
```
